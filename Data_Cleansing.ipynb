## Description

```python
wine_review_df.head()
```

```python
words_to_remove = ['and', 'much', 'like', 'the', 'from', 'this']

def remove_words(text):
    words = text.split()
    cleaned_words = [word.lower().replace(',', '') for word in words if word.lower() not in words_to_remove and len(word) > 2]
    return ' '.join(cleaned_words)

wine_review_df['description'] = wine_review_df['description'].apply(remove_words)
```

```python
wine_review_df.head()
```

# Add continent column

```python
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
        return country_continent_name
    except:
        return None
wine_review_df['continent'] = wine_review_df['country'].apply(country_to_continent)
wine_review_df.head(10)
```

```python
wine_review_df.continent.value_counts()
```

```python
wine_review_df.loc[(wine_review_df['continent'].isnull()) & (wine_review_df['country'] == 'US'), 'continent'] = 'North America'
```
